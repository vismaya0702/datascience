```python

import seaborn as sns
# Matching array lengths
bmi = np.array([1, 2, 3])
age = np.array([4, 5, 6])
sugar = np.array(['yes', 'no', 'yes'])

# Plot with class labels using hue
sns.scatterplot(x=bmi, y=age, hue=sugar, palette={'yes': 'red', 'no': 'green'})
plt.title("BMI vs Age with Sugar Label")
plt.show()
```
